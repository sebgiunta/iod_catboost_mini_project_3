## Reading Data

```python
# Import Modules

# Standard
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# CatBoost
from catboost import CatBoostClassifier, cv, Pool

# Sklearn
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import f1_score, accuracy_score, make_scorer, confusion_matrix
from sklearn.preprocessing import StandardScaler, LabelEncoder

#Imblean
from imblearn.under_sampling import ClusterCentroids

# Custom Code
from show_summary_report import show_summary_report

# Pandas Options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# Read in data

df = pd.read_csv('datasets/attrition.csv')
df.head()
```

## Data Preperation

```python
# Convert Attrition (Target Variable) columns to 0 = No & 1 = Yes

df.Attrition = df.Attrition.apply(lambda x: 0 if x == 'No' else 1)

# Convert Overtime columns to 0 = No & 1 = Yes

df.OverTime = df.OverTime.apply(lambda x: 0 if x == 'No' else 1)

# Convert travel frequency to cardinality (0,1,2)
travel_dict = {'Non-Travel': 0,
              'Travel_Rarely': 1,
              'Travel_Frequently': 2}

df = df.replace({'BusinessTravel':travel_dict})

# Drop EmployeeCount, StandardHours, Over18 - These columns are stable

df = df.drop(["EmployeeCount", "StandardHours", "Over18", "EmployeeNumber"], axis=1)

# Get Dummies
df = pd.get_dummies(df)
```

```python
# Set y and X variables

y = df.Attrition
X = df.drop('Attrition', axis=1)
```

```python
# Check Target Variable Balance

print('Labels: {}'.format(set(y)))
print('Zero count = {}, One count = {}'.format(len(y) - sum(y), sum(y)))
display(y.value_counts(normalize=True))
```

As we have data that isn't balanced, we'll be using the Recall score as our evaluation metric and will need to factor in class weights when training the model.

## Data Split

```python
# Split data into train and test

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=46, stratify=y)
```

```python
# Check Target Variable Balance after Train/Test Split

print('Labels: {}'.format(set(y_train)))
print('Zero count = {}, One count = {}'.format(len(y_train) - sum(y_train), sum(y_train)))
```

## Cross Validation

```python
# Set Parameters
params = {'loss_function': 'Logloss',
         'random_seed': 46,
         'verbose': 200,
         'early_stopping_rounds': 200,
         'eval_metric': 'Recall'}

# Create Pool class
all_train_data = Pool(data=X_train,
                      label=y_train,
                     )

# Run 5 Fold Validation
cv_scores = cv(pool=all_train_data,
            params=params, 
            fold_count=5,
            seed=46,
            shuffle=True,
            stratified=True)
```

```python
print('Best validation Recall score: {:.2f}±{:.2f} on step {}'.format(
    np.max(cv_scores['test-Recall-mean']),
    cv_scores['test-Recall-std'][np.argmax(cv_scores['test-Recall-mean'])],
    np.argmax(cv_scores['test-Recall-mean'])
))
```

```python
print('Precise validation Recall score: {}'.format(np.max(cv_scores['test-Recall-mean'])))
```

```python
print('Average of Mean Recall Score: ', np.mean(cv_scores['test-Recall-mean']))
```

From this initial check, it's clear that the imbalanced data is impacting the Recall. This'll need to be addressed.

## Model Fit - #1

The only parameter we'll include is the `early_stopping_rounds` which is Catboost overfitting detector. It'll stop iterating when it detects limited improvement in the test performance.

```python
%%time

# Set Parameters
params = {'random_seed': 46,
         'verbose': 200,
         'early_stopping_rounds': 200,
         'eval_metric': 'Recall'}

# Model
model = CatBoostClassifier(**params)

# Fit
model.fit(X_train, y_train, eval_set=(X_test, y_test), use_best_model=True)
```

```python
# Train Predictions
y_train_pred = model.predict(X_train)
y_train_prob = model.predict_proba(X_train)

# Display Train Results
show_summary_report(y_train, y_train_pred, y_train_prob)
```

```python
# Test Predictions
y_test_pred = model.predict(X_test)
y_test_prob = model.predict_proba(X_test)

# Display Test Results
show_summary_report(y_test, y_test_pred, y_test_prob)
```

There is clearly overfitting, as well as a recall of only 0.2128.

## Model Fit - #2

Given the class imbalance, we need to factor in the class weights. We'll include that on this model run.

```python
%%time

# Set Parameters
params = {'random_seed': 46,
         'verbose': 200,
         'early_stopping_rounds': 200,
         'eval_metric': 'Recall:use_weights=False',
         'class_weights': {0:1, 1:5}}

# Model
model = CatBoostClassifier(**params)

# Fit
model.fit(X_train, y_train, eval_set=(X_test, y_test), use_best_model=True)
```

```python
# Train Predictions
y_train_pred = model.predict(X_train)
y_train_prob = model.predict_proba(X_train)

# Display Train Results
show_summary_report(y_train, y_train_pred, y_train_prob)
```

```python
# Test Predictions
y_test_pred = model.predict(X_test)
y_test_prob = model.predict_proba(X_test)

# Display Test Results
show_summary_report(y_test, y_test_pred, y_test_prob)
```

We've seen an improvement on the Recall, increased from 0.2128 to 0.5745. Though it's clear this data imbalanced is still impacting the model performance.

To address this, we'll use a method called ClusterCentroids, which is an undersamping method. This method will create clusters in the majority class, of similar data points. These clusters will then be reduce to their corresponding centroid.

## Undersampling

```python
# Instantiate Cluser Centroids class
cc = ClusterCentroids()

# Resample X & y
X_cc, y_cc = cc.fit_resample(X, y)

# Check
y_cc.value_counts()
```

As a result of using this method, we now have a target class with equal number of 1 and 0.

```python
# New Train/Test Split

X_train, X_test, y_train, y_test = train_test_split(X_cc, y_cc, test_size=0.2, random_state=46)
```

## Model Fit - #3

```python
%%time

# Set Parameters
params = {'random_seed': 46,
         'verbose': 200,
         'early_stopping_rounds': 200,
         'eval_metric': 'Recall'}

# Model
model = CatBoostClassifier(**params)

# Fit
model.fit(X_train, y_train, eval_set=(X_test, y_test), use_best_model=True)
```

```python
# Train Predictions
y_train_pred = model.predict(X_train)
y_train_prob = model.predict_proba(X_train)

# Display Train Results
show_summary_report(y_train, y_train_pred, y_train_prob)
```

```python
# Test Predictions
y_test_pred = model.predict(X_test)
y_test_prob = model.predict_proba(X_test)

# Display Test Results
show_summary_report(y_test, y_test_pred, y_test_prob)
```

Using the undersampling method, then retraining the model using this new data, we've achieved fantastic scores in the test set.

*Note that this is a synthetic dataset, which explains the reason we're able to achieve such high scores*
